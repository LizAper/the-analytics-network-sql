To calculate the trip length, we could subtract the start coordinates from the destination coordinates.

```r
# We calculate the length and distance of each trip.
dataset$travel_length <- distVincentySphere(cbind(dataset$start_lat, dataset$start_lng), cbind(dataset$end_lat, dataset$end_lng))
```

#### **Hypothesis Testing: Differences in Bicycling Patterns between Annual and Casual Members: A Weekly Exploration.**

To perform a hypothesis test in R, we must first define our null and alternative hypotheses.
- **Null hypothesis**: There is no difference in the number of trips taken by annual members and occasional users during the workweek.
- **Alternative hypothesis**: Annual members take more trips during the workweek than occasional users.

We can perform a hypothesis test using a two-sample t-test:

```r
# Create a data frame with the data provided.
data <- data.frame(
  weekday = c("Thursday", "Wednesday", "Tuesday", "Monday", "Saturday", "Friday", "Saturday", "Sunday", "Sunday", "Friday", "Thursday", "Monday", "Wednesday", "Tuesday"),
  member_casual = c(rep("member", 7), rep("casual", 7)),
  number_of_trips = c(404329, 401152, 400640, 365961, 354015, 349877, 328680, 290242, 289240, 239999, 222113, 204213, 196409, 189839)
)

# separate data for annual and casual members
annual_members <- subset(data, member_casual == "member")
casual_members <- subset(data, member_casual == "casual")

# Perform hypothesis tests to compare the number of weekday and weekend trips.

# For annual members
annual_weekly_result <- t.test(annual_members$number_of_trips[annual_members$weekday %in% c("Monday", "Tuesday", "Wednesday", "Thursday", "Friday")],
                                   annual_members$number_of_trips[annual_members$weekday %in% c("Saturday", "Sunday")],
                                   paired = FALSE)

# For casual members
casual_weekly_result <- t.test(casual_members$number_of_trips[casual_members$weekday %in% c("Monday", "Tuesday", "Wednesday", "Thursday", "Friday")],
                                    casual_members$number_of_trips[casual_members$weekday %in% c("Saturday", "Sunday")],
                                    paired = FALSE)

# print the results
print("Hypothesis test for annual members:")
print(annual_weekly_result)

print("Hypothesis test for casual members:")
print(casual_weekly_result)
```

The t-test results for annual and casual members provide additional information on differences in the number of trips between weekdays and weekends.

For **annual members**:

Test statistic (t): **2.5536**.

Degrees of freedom (df): **2.73**

P-value (p-value): **0.09192**.

**Interpretation:**

The p-value (0.09192) is greater than the common significance level of 0.05. This suggests that there is insufficient evidence to reject the null hypothesis. In other words, there is no significant difference in the number of weekday and weekend trips for annual members according to the t-test.

For **casual members:**

Test statistic (t): **-8.6571.**

Degrees of freedom (df): **4.0239**

p-value (p-value): **0.0009536.**

**Interpretation:**

The p-value (0.0009536) is less than 0.05, indicating that there is sufficient evidence to reject the null hypothesis. In other words, there are significant differences in the number of weekday and weekend trips for casual members according to the t-test.

These results are consistent with those obtained earlier. The t-test for annual members suggests that there are no significant differences, while for casual members it indicates significant differences in the number of weekday and weekend trips. In summary, the theory that casual members might use bicycles more on weekends continues to be supported by the evidence.

### **Hypothesis test: Differences in the type of bicycle used by annual members and occasional users.**

To conduct a hypothesis test comparing bicycle preferences between annual and occasional members, we can perform a chi-square test to assess whether there is a significant association between membership type and type of bicycle used.

**Null Hypothesis (H0):** There is no significant association between the type of bicycle used and the user's membership type.

**Alternative Hypothesis (H1):** There is a significant association between the type of bicycle used and the user's membership type. Bicycle preferences are not independent of membership type.

```r
# Create a contingency table
contingency_table <- table(dataset$rideable_type, dataset$member_casual)

# Perform the chi-square test
chi_square_test <- chisq.test(contingency_table)
print(chi_square_test)
```

- p-value very close to zero (p-value < 2.2e-16). This indicates that there is a significant association between bicycle type and membership type.

Since the p-value is extremely small, we reject the null hypothesis and conclude that there are significant differences in bicycle preference between the two membership groups.
